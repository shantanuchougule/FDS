```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = ('Kolhapur', 'Sangli', 'Gadhinglaj', 'Miraj', 'Gargoti', 'Solapur','Mumbai')
y = ('22', '25', '28', '32', '35', '38', '40')
plt.plot(x, y)
plt.title("Temperature Graph")
plt.show()
```

```python
#PI GRAPH
x = ('Kolhapur', 'Sangli', 'Gadhinglaj', 'Miraj', 'Gargoti', 'Solapur','Mumbai')
y = (22, 25, 28, 32, 35, 38, 40)
plt.pie(y,labels = x, autopct = '%1.1f%%')
plt.title("Temperature Graph")
plt.show()
```

```python
#BAR GRAPH
plt.bar(x,y)
plt.title("Temperature Graph")
plt.xlabel("Cities")
plt.ylabel("Temperature")
plt.show()
```

```python
#SCATTER GRAPH
plt.scatter(x,y)
plt.title("Temperature Graph")
plt.xlabel("Cities")
plt.ylabel("Temperature")
plt.show()
```
