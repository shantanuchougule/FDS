```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python
#Loading the dataset
df =pd.read_csv("bombay.csv")
df.head()
```

# 1. Normal Distribution:

```python
sns.kdeplot(data=df['humidity'])
plt.xlabel('Range of humidity')
plt.ylabel('Density')
plt.title('Normal Distribution Plot of Bombay temperature')
plt.show()
```

# 2. Poisson Distribution:

```python
lambda_av = df["maxtempC"].mean()
print(lambda_av)

n_points = 1000
poisson_data = np.random.poisson(lam=lambda_av, size=n_points)

sns.displot(poisson_data, kde=False, color='purple')
plt.xlabel('Maximum Temperature')
plt.ylabel('Frequency')
plt.title('Poisson Distribution of maxtemp in Bombay ')
```

# 3. Uniform Distribution

```python
sns.distplot(np.random.uniform(size=5000), hist=True)
```

# 4. Gamma Distribution

```python
#Importing the required library 

from scipy.stats import gamma

data_column = "pressure"

shape, loc, scale = gamma.fit(df[data_column])

x = np.linspace(min(df[data_column]), max(df[data_column]), 100)
gamma_pdf = gamma.pdf(x, shape, loc=loc, scale=scale)

sns.kdeplot(df[data_column], shade=True, label="Data Distribution", color = 'green') 
plt.plot(x, gamma_pdf, label="Fitted Gamma Distribution")
plt.xlabel(data_column)
plt.ylabel("Density") 
plt.title("Gamma Distribution Fit to " + data_column)
plt.legend() 
plt.show()
```
